```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('train.csv')
#df=pd.read_csv('test.csv')
df.columns
```

```python
df.head()
```

```python
df.head()
```

```python
del df["Name"]
del df["Ticket"]
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
del df["Cabin"]
df.head()
```

```python
MeanS=df[df.Survived==1].Age.mean()
MeanNS=df[df.Survived==0].Age.mean()
```

```python
df['age']=np.where(df.Age.isnull() & df["Survived"]==1,MeanS,df["Age"])
df.head()
```

```python
df.age.fillna(MeanNS,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
del df['Age']
df.head()
```

```python
df.rename(columns={"age":"Age"}, inplace=True)
df.head()
```

```python
def gender(a):
    if a=="male":
        return 2
    else:
        return 1
df["Sex"]=df["Sex"].apply(gender)
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
def embarked(a):
    if a=="S":
        return 1
    elif a=="C":
        return 2
    else:
        return 3
df["Embarked"]=df["Embarked"].apply(embarked)
df.head()
```

```python
df.shape
df = df[['Pclass', 'Sex',"SibSp",'Parch',"Fare","Embarked","Age","Survived"]]
df.head()
```

```python
X=df.iloc[:,0:7]
y=df.iloc[:,7]
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
y_train.shape
y_train.head()
```

```python
#clf=LogisticRegression(C=1)
clf=KNeighborsClassifier(n_neighbors=3)
clf.fit(X,y)
```

```python
predict=clf.predict(X_test)
predict
```

```python
score=clf.score(X_test,y_test)
score
```
